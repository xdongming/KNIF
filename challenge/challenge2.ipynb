```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
from itertools import product
from matplotlib import pyplot as plt
import math
```

```python
from matplotlib.font_manager import FontProperties

arial_font = FontProperties(fname="/root/.fonts/arial.ttf", size=20)

plt.figure(figsize=(5, 4))
n_values = np.arange(1, 6)
y_values = [math.comb(n + 3, n) for n in n_values]
plt.plot(n_values, y_values, color='red', marker='o')
plt.yticks(fontproperties=arial_font)
plt.xticks((1,2,3,4,5), fontproperties=arial_font)
plt.xlabel('Order', fontproperties=arial_font)
plt.ylabel('Number', fontproperties=arial_font)
# plt.axis('off')
plt.tight_layout()
plt.savefig("b3.svg", format='svg', bbox_inches='tight')
plt.show()
```

```python
mse_edmddl = np.load('mse_edmddl.npy')
mse_knif = np.load('mse_knif.npy')
```

```python
from matplotlib.font_manager import FontProperties

t = np.linspace(0.01, 0.6, 60)

arial_font = FontProperties(fname="/root/.fonts/arial.ttf", size=20)

plt.figure(figsize=(5, 4))
plt.plot(t, mse_knif, color='blue', label='Ours')
plt.plot(t, mse_edmddl, color='red', label='Baseline')
plt.yticks(fontproperties=arial_font)
plt.xticks(fontproperties=arial_font)
plt.yscale('log')
plt.legend(
    loc='lower right',
    prop=arial_font
)
plt.xlabel('$t$', fontproperties=arial_font)
plt.ylabel('Reconstruction MSE', fontproperties=arial_font)
# plt.title(r'Plot of $C_{n+3}^n$ vs $n$')
# plt.axis('off')
plt.tight_layout()
plt.savefig("b4.svg", format='svg', bbox_inches='tight')
plt.show()
```
